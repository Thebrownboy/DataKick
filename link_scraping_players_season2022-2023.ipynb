```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
from selenium import webdriver
import time 

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

import pandas as pd 
import numpy as np 

season_link = "https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures"
browser= web_driver()


browser.set_page_load_timeout(15)


try :
    browser.get(season_link)
    # just to catch the timeout error 
except:
    pass

browser.set_page_load_timeout(10)

match_reports_links=browser.find_elements(By.CSS_SELECTOR,"td[data-stat='match_report']>a")

match_reports_links_lst=[]

for i in range(len(match_reports_links)):
    match_reports_links_lst.append(match_reports_links[i].get_attribute("href"))
  
#
data_frame={"Player":[],"shirt_number":[],"Nation":[],"Pos":[],"Age":[],"Min":[],"Gls":[],
            "Ast":[],	"PK":[],	"PKatt":[],	"Sh":[],	"SoT":[],	"CrdY":[],	"CrdR":[],	"Touches":[],	"Tkl":[],	"Int":[],	"Blocks":[]}


# for match_item_link in match_reports_links_lst:
def preprocessing(text_table):
  # remove first two rows
  text_table = text_table[text_table.find('\n')+1:]
  text_table = text_table[text_table.find('\n')+1:]
  # remove last row
  text_table = text_table[:text_table.rfind('\n')]
  idx=0
  for line in text_table.split('\n'):
    name = ""
    for c in line:
      if c.isdigit():
        break
      else:
          name=name+c
    other = line[len(name):].split()
    other.insert(0,name)
    other.pop(2)
    for i, key in enumerate(data_frame.keys()):
        if i<len(other):
          data_frame[key].append(other[i])
    idx=idx+1

print(len(match_reports_links_lst))

browser.set_page_load_timeout(50)

for idx,match_item_link in enumerate(match_reports_links_lst):
  print(match_reports_links_lst[idx])
  try:
        browser.get(match_reports_links_lst[idx])
        text_table=browser.find_element(By.XPATH,"/html/body/div[2]/div[5]/div[8]/div[8]/div[1]/table").text
        preprocessing(text_table)
  except:
        print("pass")
        pass

  try:
        browser.get(match_reports_links_lst[idx])
        text_table=browser.find_element(By.XPATH,"/html/body/div[2]/div[5]/div[11]/div[8]/div[1]/table").text
        preprocessing(text_table)
  except:
        print("pass")
        pass
  

#print(data_frame)
df=pd.DataFrame(data_frame)    
df.to_csv("temp2.csv")
```
