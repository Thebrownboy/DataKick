```python
import pandas as pd 
import numpy as np 
import os 
import subprocess 
from  matplotlib import pyplot as plt  
```

```python
df  = pd.read_csv("Formation_Data/Liverpool.csv")
```

```python
opponent_lst = os.listdir("Formation_Data")
opponent_lst.remove('Liverpool.csv')
```

```python
opponent = pd.read_csv("Formation_Data/West Ham.csv")
opponent
```

```python
formation_list= [ ]
index = 0 
for opponent in opponent_lst :
    opponent = pd.read_csv(f"Formation_Data/{opponent}")
    opponent_query_result = opponent.query("Opponent =='Liverpool' and  (GF > GA or GF==GA )")
    if (opponent_query_result.count()[0]>0):
        for i in range(opponent_query_result.count()[0]):
            formation_list.append(opponent_query_result.iloc[i]["Formation"])
               
```

```python

counts= {i: formation_list.count(i) for i in formation_list}
```

```python
labels= counts.keys()
counts=counts.values()
```

```python
ax = plt.bar(labels,counts,)
plt.xlabel("Formation")
plt.ylabel("Number of Negative Result")
plt.title("number of times the team loses against specific formation")
plt.savefig("../screenshots/formation_stats")
plt.show()
```
