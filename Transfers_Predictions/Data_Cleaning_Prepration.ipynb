```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import pickle 
pd.set_option("display.max_columns",90)
```

```python
season  = "2015-2016"
season_1= "2016-2017"
names=None
with open(f"sstand-stats{season}/season-{season}","rb") as fb : 
    names = pickle.load(fb)
print(names)

nones=names[-3:]
names= names[0:-3]
```

```python
train_data=pd.DataFrame()
```

```python
for i , name in enumerate(names):
    print(name)
    team_df = pd.read_csv(f"sstand-stats{season}/{name}-stand-stats.csv")
    trans_df = pd.read_csv(f"sstand-stats{season}/transfers-{season_1}.csv")
    team_df =team_df[team_df.columns[:-2]]
    team_df.drop(["Player","Nation","CrdY","CrdR"],axis=1,inplace=True)
    team_df= team_df.iloc[:-2]
    team_df["Pos"]= team_df["Pos"].apply(lambda x : x[0:2])
    team_df= team_df[team_df["MP"]>0] #remove the players that didn't play in the season
    
    team_df= team_df[team_df.columns[:-5]]
    team_df= pd.concat([team_df,pd.DataFrame({"rank":[i+1]*team_df.shape[0]})],axis=1)
    num_of_players=team_df[team_df["Pos"]=="FW"].shape[0]
    obs=pd.DataFrame(team_df[team_df["Pos"]=="FW"].describe().loc["mean"]).T.reset_index()
    
   
    obs=pd.concat([obs,pd.DataFrame({"team_name":[name]})],axis=1)
    obs=pd.concat([obs,pd.DataFrame({"#attackers":[num_of_players]})],axis=1)
#     team_df= pd.concat([team_df,pd.DataFrame({"team_name":[name]*team_df.shape[0]})],axis=1)
    train_data=pd.concat([train_data,
                          obs]
                         ,axis=0)
    
    
    
    
```

```python


train_data= train_data.reset_index().drop(["level_0","index"],axis=1)
```

```python
train_data["team_name"]=train_data["team_name"].apply(lambda x : x.replace("Utd","United"))
train_data["team_name"]=train_data["team_name"].apply(str.lower)
train_data["team_name"]=train_data["team_name"].apply(lambda x : x.replace(" ",""))
train_data.sort_values(by=["team_name"],ascending=True,inplace=True)
train_data=train_data.reset_index().drop(["index"],axis=1)
```

```python
transfers = pd.read_csv(f"sstand-stats{season}/transfers-{season_1}.csv")
transfers.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
transfers
```

```python
transfers["team"]=transfers["team"].apply(lambda x : x.replace("REAL VALLADOLID CF","VALLADOLID"))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("AFC",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("FC",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("CA",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("CF",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("RCD",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("CD",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("DEPORTIVO",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("SD",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("UD",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("DE",""))






transfers["team"]=transfers["team"].apply(str.lower)
transfers["team"]=transfers["team"].apply(lambda x : x.replace(" ",""))
transfers["team"]=transfers["team"].apply(lambda x : x.replace("wolverhamptonwanderers","wolves"))


```

```python
transfers
```

```python
transfers["Attacking_transfers"]=transfers["RightWinger"] + transfers["LeftWinger"] \
+ transfers["SecondStriker"]+  transfers["Centre-Forward"]
```

```python
transfers= transfers[["team","Attacking_transfers"]]
transfers["team"]= transfers["team"].replace(['realbetisbalompié'],["betis"])
transfers["team"]= transfers["team"].replace(['rcdespanyolbarcelona'],["espanyol"])


transfers= transfers.sort_values(by="team",ascending=True).reset_index().drop("index",axis=True)
```
