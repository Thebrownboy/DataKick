```python
train_data=pd.DataFrame()
```

```python
for i , name in enumerate(names):
    team_df = pd.read_csv(f"stand-stats2021-2022/{name}-stand-stats.csv")
    trans_df = pd.read_csv("stand-stats2021-2022/transfers-2022-2023.csv")
    team_df =team_df[team_df.columns[:-2]]
    team_df.drop(["Player","Nation","CrdY","CrdR"],axis=1,inplace=True)
    team_df= team_df.iloc[:-2]
    team_df["Pos"]= team_df["Pos"].apply(lambda x : x[0:2])
    team_df= team_df[~team_df["Min"].isna()] #remove the players that didn't play in the season
    team_df= team_df[team_df.columns[:-10]]
    team_df= pd.concat([team_df,pd.DataFrame({"rank":[i]*team_df.shape[0]})],axis=1)
#     team_df= pd.concat([team_df,pd.DataFrame({"team_name":[name]*team_df.shape[0]})],axis=1)
    train_data=pd.concat([train_data,
                          pd.DataFrame(team_df[team_df["Pos"]=="FW"].describe().loc["mean"]).T]
                         ,axis=0)
```
